## Import packages and upload the database

```python
import pandas as pd
from scipy.stats import mstats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sqlalchemy import create_engine
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
```

```python
# Database connection
engine = create_engine('postgresql://admin:password@localhost:5433/heartdisease')
query = "SELECT age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal FROM patient_data"
df = pd.read_sql(query, engine)
```

## Clean each categorical column and impute missing values using KNN

```python
# Define valid values for each categorical column
valid_values = {
    'sex': [0, 1],
    'cp': [1, 2, 3, 4],
    'fbs': [0, 1],
    'restecg': [0, 1, 2],
    'exang': [0, 1],
    'slope': [1, 2, 3],
    'thal': [3, 6, 7],
    'ca': [0, 1, 2, 3]
}
print("Distribution of data after removing invalid values:")

for column, valids in valid_values.items():
    df[column] = df[column].where(df[column].isin(valids))
    print(df[column].value_counts(dropna=False))

# Impute missing values using KNN on scaled data
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
imputer = KNNImputer(n_neighbors=5)
df_imputed_scaled = imputer.fit_transform(df_scaled)

# Convert scaled imputed data back to DataFrame
df_imputed = pd.DataFrame(scaler.inverse_transform(df_imputed_scaled), columns=df.columns)

# Snap to nearest valid category
for column in valid_values:
    unique_values = np.array(valid_values[column])
    df_imputed[column] = df_imputed[column].apply(lambda x: unique_values[np.argmin(np.abs(unique_values - x))])

# Ensure the data types are correct
for column in df.columns:
    if column in valid_values:
        df_imputed[column] = df_imputed[column].astype(int)

# Print final distribution
print("Final data distribution after KNN imputation and adjustment:")
for column in valid_values:
    print(f"{column} distribution after imputation and adjustment:")
    print(df_imputed[column].value_counts())

for column in valid_values:
    df[column] = df_imputed[column]
```

## Justifying Choices

### Justification for Using KNN Imputation in Heart Disease Dataset

#### **1. Handling Multidimensional Data**
KNN (K-Nearest Neighbors) imputation is particularly effective for datasets where the relationship between variables is crucial. In the context of heart disease data, variables are likely interdependent, influencing the development and indicators of the disease. KNN considers these relationships by using the similarity between samples (patients) to impute missing values, preserving the inherent structure of the data.

#### **2. Capability to Impute Categorical Data**
While KNN is commonly used for continuous data, it can be adapted for categorical data through appropriate preprocessing steps like scaling. By finding the nearest neighbors in a multidimensional space, KNN can effectively estimate the most plausible categories for missing data points, based on the distribution and frequency of those categories among the closest samples.

#### **3. Flexibility and Accuracy**
KNN can adapt to the specific data structure better than simpler imputation methods like mean or median imputation, which do not account for variability in data relationships. It uses a non-parametric method, making fewer assumptions about the data distribution and thus can provide more accurate imputations for complex datasets like those involving health data.

#### **4. Ease of Implementation**
With libraries such as Scikit-learn providing straightforward implementations of KNN imputation, it is accessible and relatively easy to integrate into a data preprocessing pipeline, offering a good balance between complexity and performance.

### Conclusion
Given the complex relationships in heart disease data and the importance of accurately handling missing values without distorting underlying patterns, KNN imputation is a suitable choice. It leverages the dataset's multidimensional nature to provide more reliable and meaningful imputations compared to simpler, assumption-heavy methods.

## Handle numerical columns outliers using quantile method 

```python
upper_limit_chol = df['chol'].quantile(0.95)
upper_limit_trestbps = df['trestbps'].quantile(0.95)
upper_limit_thalach = df['thalach'].quantile(0.95)

zeros_chol = (df['chol'] == 0).sum()
df['chol'] = df['chol'].replace(0, np.nan)

# Define the lower limit for capping 'thalach' if needed, considering physiological limits
lower_limit_thalach = df['thalach'].quantile(0.05)

# Cap 'chol', 'trestbps' values at their 95th percentile
df['chol'] = np.where(df['chol'] > upper_limit_chol, upper_limit_chol, df['chol'])
df['trestbps'] = np.where(df['trestbps'] > upper_limit_trestbps, upper_limit_trestbps, df['trestbps'])
df.loc[df['chol'] < 150, 'chol'] = np.nan
# Cap 'thalach' values at their 95th percentile and at the 5th percentile if needed
df['thalach'] = np.where(df['thalach'] > upper_limit_thalach, upper_limit_thalach, df['thalach'])
df['thalach'] = np.where(df['thalach'] < lower_limit_thalach, lower_limit_thalach, df['thalach'])
```

## Justifying Choices

### Justification for Using Quartiles to Remove Outliers in Heart Disease Dataset

#### **1. Robustness to Extreme Values**
Quartiles are particularly effective in managing outliers because they focus on the central distribution of the data, reducing the influence of extreme values. By capping values at the 5th and 95th percentiles, you effectively limit the impact of outliers which are significantly higher or lower than the typical range. This is crucial in clinical data like heart disease metrics, where extreme values can skew analysis and might not represent typical patient conditions.

#### **2. Preserving Data Integrity**
Using the 95th percentile to cap values like 'chol' (cholesterol) and 'trestbps' (resting blood pressure) ensures that the bulk of the data remains untouched, preserving the overall integrity and distribution of the dataset. This method avoids the loss of essential data by not removing outlier observations entirely but rather reducing their influence.

#### **3. Handling Biologically Implausible Values**
Setting zeros in 'chol' to NaN is a specific decision to address biologically implausible values. Cholesterol levels of zero are not possible physiologically and likely represent missing or incorrectly entered data. By replacing these zeros with NaN, you prepare the dataset for more accurate imputation methods.

#### **4. Enhancing Model Performance**
Capping and flooring the data using percentile values can lead to improved performance in predictive modeling. Models often perform better when extreme outliers are controlled, as these can disproportionately influence the model's training phase, leading to overfitting or poor generalization to new data.

#### **5. Tailoring to Clinical Relevance**
In clinical settings, understanding the typical range and slight deviations is often more valuable than extreme values, which could be anomalies. By focusing on the 5th to 95th percentile range, your analysis will better reflect conditions that are clinically relevant, aiding in more accurate diagnoses or predictions related to heart disease.

### Conclusion
The use of quartiles for managing outliers in the heart disease dataset is a statistically sound and practically effective method. It maintains the quality of the data while ensuring that the analysis is not compromised by extreme values, making subsequent statistical analysis and modeling more reliable.


## Handle missing values using mean/median for the numerical columns

```python
# Impute missing values using mean for age and cholesterol and convert to integer
df['age'] = df['age'].fillna(df['age'].mean()).astype(int)

# Impute missing values using median for trestbps and thalach and convert to integer
df['thalach'] = df['thalach'].fillna(df['thalach'].mean()).astype(int)
#df['chol'] = df['chol'].fillna(df['chol'].mean()).astype(int)


# Since 'oldpeak' can be zero and may include negative or positive shifts, median is used and converted to integer
df['oldpeak'] = df['oldpeak'].fillna(df['oldpeak'].mean()).astype(float)

# Winsorize 'trestbps' and 'chol' before imputation
df['trestbps'] = mstats.winsorize(df['trestbps'], limits=[0.05, 0.05])

# Apply median imputation for missing values
imputer = SimpleImputer(strategy='mean')
df['trestbps'] = imputer.fit_transform(df[['trestbps']])

# Convert columns back to integer if they are supposed to be integers
df['trestbps'] = df['trestbps'].astype(int)


df['chol'] = df['chol'].fillna(df['chol'].mean()).astype(int)
```

## Justifying Choices

### Justification for Using Mean and Median to Handle Missing Numerical Data

#### **1. Imputing with Mean**
- **Use Case:** The mean is a common measure of central tendency that is appropriate to use when the data distribution is symmetrical or when outliers are minimal. It provides a good estimate of the "central" value for a set of data.
- **Example:** Imputing age and cholesterol with the mean in the heart disease dataset makes sense if these variables are normally distributed without extreme outliers. By filling in missing values with the mean, we assume that the missing values are randomly distributed around the central tendency.
- **Conversion to Integer:** Converting these values to integers after imputation makes sense if the data schema requires these fields to be integers, perhaps for consistency in reporting or data storage.

#### **2. Imputing with Median**
- **Use Case:** The median is the middle value of a data set when ordered and is less affected by outliers and skewed data than the mean. It is typically used for data with an asymmetrical distribution or when outliers are present, providing a better "central" measure that is representative of typical data points.
- **Example:** For variables like 'trestbps' (resting blood pressure) and 'thalach' (maximum heart rate), which can be affected by extreme values or a skewed distribution, using the median is a reliable choice. It ensures that the imputed value is not influenced by outliers, which might distort the analysis if the mean were used.

#### **3. Special Case for 'oldpeak':**
- **Rationale:** The 'oldpeak' variable, indicating ST depression induced by exercise relative to rest, could have a range that includes zero and potentially negative or positive values. Imputing missing values for such a variable with the median is often more appropriate due to the potential skew in its distribution. However, there is a mention of using the mean in the example which could be a typo or oversight. Ideally, median imputation should be used here if the distribution is skewed.

#### **4. Winsorizing Before Imputation:**
- **Rationale:** Winsorizing 'trestbps' and 'chol' (capping the extreme percentiles of the data) before imputation helps in reducing the influence of extreme outliers. This step makes the subsequent mean or median imputation more effective and representative of the central tendency of the general population.

### Conclusion
Choosing the appropriate method for imputing missing values—mean or median—depends significantly on the characteristics of the data distribution for each variable. By carefully selecting these methods based on the data's nature, the integrity and accuracy of the dataset are maintained, ensuring more reliable analyses and outcomes.
